b) Using the given the matrix object `P` in the cell below, representing the transition matrix
$$
\begin{pmatrix}
0.5 & 0.5 \\
0.2 & 0.8
\end{pmatrix},
$$
write a function to calculate the probabilities $r_{ij}(n)$ for the four possible cases after $n$ steps.

```python
import numpy as np

def r(P,n):
    return np.linalg.matrix_power(P,n)
P = [
[0.5, 0.5],
[0.2, 0.8]
]
```

c) Plot the evolution of $r_{ij}(n)$ against $n$ for 10 steps for all four combinations of $i$ and $j$. Make sure to label the axes.

*Hint*: you can use `pyplot` from the `matplotib` package to create plots in Jupyter
https://matplotlib.org/tutorials/introductory/pyplot.html

```python
n_values = [i for i in range(11)]
r00 =[]
r01 =[]
r10 =[]
r11 =[]
for n in n_values:
    rr= r(P,n)
    r00.append(rr[0][0])
    r01.append(rr[0][1])
    r10.append(rr[1][0])
    r11.append(rr[1][1])
```

```python
import matplotlib.pyplot as plt

fig= plt.figure(figsize=(15,6))
ax = fig.add_subplot(111)
ax.plot(n_values, r00, label ='r00')
ax.plot(n_values, r01, label ='r01')
ax.plot(n_values, r10, label ='r10')
ax.plot(n_values, r11, label ='r11')
ax.set_xlabel('n')
ax.set_ylabel('r')
ax.set_title('Evolution of R against n')
plt.legend()
```
